```python
# for loop

cities = ["Austin", "Allen", "Dallas", "Frisco", "New York", "Rhode", "Bryan", "San Diego"]

new_cities = []

for x in cities:
    if x[0] == "A":
        new_cities.append(x)
        
print(new_cities)
```

```python
numbers = [6,4,12,9,3,2,11]

new_numbers = []

for i in numbers:
    if i >=6:
        new_numbers.append(i)
        
print(new_numbers)
```

```python
# dataframe to show different things

file = pd.read_csv("cereal.csv")

cereal_df = pd.DataFrame(file)

cereal_df.head(5)
```

```python
# describe 

cereal_df.describe()
```

```python
# formatting 

cereal_df["rating"]= cereal_df["rating"].map("{:.2f}".format)
```

```python
cereal_df.head(2)
```

```python
# set index

cereal_df = cereal_df.set_index("name")
```

```python
cereal_df.head(3)
```

```python
# sort_values - dataframe

cereal_df.sort_values(by=["calories"],ascending=False).head()
```

```python
# iterrows():

for i, row in cereal_df.iterrows():
    if row[2] >= 120:
        print(i, " : ", row[2])
```

```python
# using list compression to get a list from a DF

calories = [x for x in cereal_df["calories"]]
len(calories)
```

```python
# import new dataset 

nba_file = pd.read_csv("nba.csv")
nba_df = pd.DataFrame(nba_file)
nba_df.head(5)
```

```python
nba_df["Salary"] = nba_df["Salary"].map("{:,.2f}".format)
```

```python
# loc filtering rows and selecting columns by label (row-index, column-column)

nba_df.loc[0,:]
```

```python
nba_df.loc[0:2, :]
```

```python
nba_df.loc[0:2, "Position"]
```

```python
nba_df.loc[0:5, "Name":"Position"]
```

```python
# using .loc[] with booleans

nba_df.loc[nba_df.Position == "PG", :]
```

```python
# iterrows():

pg_list = []

for i, row in nba_df.iterrows():
    if row[3] == "PG":
        pg_list.append(row[2])

print(pg_list)
```
